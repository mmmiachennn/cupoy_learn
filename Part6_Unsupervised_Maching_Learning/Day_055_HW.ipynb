- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
clu3 = KMeans(n_clusters=3).fit_predict(X)
clu6 = KMeans(n_clusters=6).fit_predict(X)
```

```python
color = {0: 'blue', 1: 'red', 2: 'orange', 3: 'black', 4: 'green', 5: 'purple'}
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in clu3])
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in clu6])
plt.show()
```

```python
clu_init5 = KMeans(n_clusters=3, n_init=5, init='random').fit_predict(X)
clu_init2 = KMeans(n_clusters=3, n_init=2, init='random').fit_predict(X)
```

```python
color = {0: 'blue', 1: 'red', 2: 'orange', 3: 'black', 4: 'green', 5: 'purple'}
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in clu_init5])
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in clu_init2])
plt.show()
```
