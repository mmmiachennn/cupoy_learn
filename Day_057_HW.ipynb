- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np
```

```python
h_w = AgglomerativeClustering(n_clusters=3, linkage="ward").fit_predict(X)
h_c = AgglomerativeClustering(n_clusters=3, linkage="complete").fit_predict(X)
h_a = AgglomerativeClustering(n_clusters=3, linkage="average").fit_predict(X)
h_s = AgglomerativeClustering(n_clusters=3, linkage="single").fit_predict(X)
```

```python
color = {0: 'blue', 1: 'red', 2: 'orange', 3: 'black', 4: 'green', 5: 'purple'}
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in h_w])
plt.title('ward')
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in h_c])
plt.title('complete')
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in h_a])
plt.title('average')
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=[color[c] for c in h_s])
plt.title('single')
plt.show()
```
